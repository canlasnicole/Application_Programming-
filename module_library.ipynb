```python
%%writefile library_system.py

# Library Constants
LIBRARY_NAME = "Central Archive"
MAX_BORROW_DAYS = 14

# Basic Library Functions
def check_out(book_title, user_name):
    return f"Book '{book_title}' has been checked out by {user_name}."

def return_book(book_title):
    return f"Book '{book_title}' is now back in the stacks."

# Library Management Class
class BookManager:
    def __init__(self, title, author):
        self.title = title
        self.author = author
        self.is_available = True

    def get_info(self):
        status = "Available" if self.is_available else "On Loan"
        return f"{self.title} by {self.author} [{status}]"

    def toggle_status(self):
        self.is_available = not self.is_available
```

```python
# 1. Defining our Library Book class (This is your 'class int')
class Book:
    def __init__(self, title):
        self.title = title
        self.is_loaned = False

    def toggle_loan(self):
        self.is_loaned = not self.is_loaned
        return True

# --- NEW CELL START ---

# Create a book instance (Like your x = 1)
my_book = Book("The Great Gatsby")

# 1. print(type(x)) -> Checking the category
print(f"Type: {type(my_book)}")

# 2. x.is_integer -> Checking a property (no parentheses)
print(f"Property Reference: {my_book.is_loaned}")

# 3. x.is_integer() -> Calling a method (with parentheses)
print(f"Method Call (Loan Toggle): {my_book.toggle_loan()}")

# 4. dir(x) -> Seeing everything a Book can do
print("\n--- Book Directory (Methods) ---")
print([item for item in dir(my_book) if not item.startswith('__')])

# 5. The hidden 'dunder' methods (Like your x.__add__(10))
# In a library, we might use __str__ to print the book name nicely
print(f"Internal Title Store: {my_book.title}")
```
