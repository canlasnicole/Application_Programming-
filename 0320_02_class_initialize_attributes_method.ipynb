## **과제 2. Tom이 Jerry를 쫒아가는 행위를 method(함수)로 만들어 보자(10번 following)**

### **class User:**
> **def __init __ (self):**
>   - ##### Initialize attributes
>   - **HAS (Attiributes)**

> **def following_mode():**
>   - ##### change attributes (following, follower) for chasing
>   - **Does (Method, Change the attributes)**

<img src= 'https://pbs.twimg.com/profile_images/810136393/tom-jerry-pic_400x400.jpg' width=300>

```python
!pip install faker

import pandas as pd
from faker import Faker
import random

fake = Faker()

class User:
    def __init__(self):
        self.name = fake.name()
        self.age = random.randint(20, 60)
        self.address = fake.address().replace("\n", " ")
        self.phone = fake.phone_number()
        self.job = fake.job()
        self.follow_count = 0

    def following_mode(self, target_name):
        self.follow_count += 1
        print(f" {self.name} is chasing {target_name}! (Attempt {self.follow_count}/10)")


user_objects = [User() for _ in range(1000)]
data_for_pandas = [vars(u) for u in user_objects]

df = pd.DataFrame(data_for_pandas)

df_final = df[['name', 'age', 'address', 'phone', 'job']]

print("# Task 1: First 5 rows of the 1,000 users DataFrame ")
display(df_final.head())
print(f"\nTotal rows generated: {len(df_final)}")
print("-" * 50)
```

```python

print(" Task 2: Tom chasing Jerry logic ")
tom = User()
tom.name = "Tom"

for _ in range(10):
    tom.following_mode("Jerry")
```
