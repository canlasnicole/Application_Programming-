```
import requests

urls = {
    "Google": "https://www.google.com",
    "Naver": "https://www.naver.com"
}

for name, url in urls.items():
    try:
        # We add a timeout so the script doesn't hang forever
        response = requests.get(url, timeout=5)

        # This will jump to the 'except' block if the status is 4xx or 5xx
        response.raise_for_status()

        print(f"{name} Headers: {response.headers.get('Content-Type')}")

    except requests.exceptions.HTTPError as http_err:
        print(f"HTTP error occurred for {name}: {http_err}")
    except requests.exceptions.ConnectionError:
        print(f"Failed to connect to {name}. Check your internet.")
    except requests.exceptions.Timeout:
        print(f"The request for {name} timed out.")
    except Exception as err:
        print(f"An unexpected error occurred for {name}: {err}")
```

```
import gradio as gr

def filter_posts(user_id):
    url = "https://jsonplaceholder.typicode.com/posts"
    # Use params to filter directly from the web request
    params = {'userId': user_id}
    res = requests.get(url, params=params)
    return pd.DataFrame(res.json())

demo = gr.Interface(fn=filter_posts, inputs="number", outputs="dataframe")
demo.launch(share=True) # share=True gives you a public link
```

```
import requests
import pandas as pd

# 1. Fetch user data from the JSONPlaceholder URL
url = "https://jsonplaceholder.typicode.com/users"
response = requests.get(url)
data = response.json()

# 2. Convert initial data into a Pandas DataFrame
df = pd.DataFrame(data)

# 3. Clean the "address" column using json_normalize()
# This turns the 'address' dictionary into a separate table with street, suite, city, etc.
address_df = pd.json_normalize(df['address'])

# 4. BONUS: Clean the "company" column too! [cite: 2]
# The company column is also 'messy' (nested). Let's clean it just like the address.
company_df = pd.json_normalize(df['company'])

# Display the results
print("--- Cleaned Address Table ---")
print(address_df.head())

print("\n--- Cleaned Company Table ---")
print(company_df.head())
```

```
import gradio as gr
import pandas as pd
import requests

def search_user(user_id):
    # 1. Fetch the data
    url = "https://jsonplaceholder.typicode.com/users"
    data = requests.get(url).json()
    df = pd.DataFrame(data)

    # 2. Filter for the ID the user typed
    # We use .iloc because IDs 1-10 correspond to rows 0-9
    user_data = df[df['id'] == int(user_id)]

    # 3. Clean up the company info for the display
    company_info = pd.json_normalize(user_data['company'])

    # 4. Create a nice summary table
    result = user_data[['id', 'name', 'email']].copy()
    result['Company Name'] = company_info['name'].values[0]

    return result

# 5. Launch the Gradio Interface
demo = gr.Interface(
    fn=search_user,
    inputs=gr.Number(label="Enter User ID (1-10)"),
    outputs="dataframe",
    title="User Directory Search"
)

demo.launch()
```
